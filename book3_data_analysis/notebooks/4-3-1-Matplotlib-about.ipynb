# 4.3.1 Matplotlibの概要

## Matplotlib とは

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.style

# ggplotスタイルを設定
matplotlib.style.use('ggplot')
```

## MATLABスタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y)  # 折れ線グラフを描画
plt.title('MATLAB-style')  # グラフにタイトルを設定

plt.show()  # グラフを表示
```

## オブジェクト指向スタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title('OOP-style')  # タイトルを設定

plt.show()  # グラフを表示
```
