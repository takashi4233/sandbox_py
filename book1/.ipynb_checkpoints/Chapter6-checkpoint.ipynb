## List,Dict,Tupleの処理一覧
|型|定義|値の取得|追加|結合|更新|削除|
|:--|:--|:--|:--|:--|:--|:--|
|list|l1=['list1','list2]|l1[0]|l1.append('list3')|l1 + l2|l1[1]='New_List1'|l1.remove('list1') |
|dict|d1={'Key1':1,'Key2':2}|d1.get('Key1')|d1.update({'Key3':100})|d1.update(d2)|d1['Key1'] = 3|del d1['Key1']|
|tuple|t1 = ('t1','t2')|t[0]| NaN|NaN|NaN|NaN|
