```python
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as random
%matplotlib inline
```

```python
#散布図
random.seed(0)
x = np.random.randn(30)
y = np.sin(x) + np.random.randn(30)

plt.figure(figsize=(20,6))
plt.plot(x,y,'o')
plt.scatter(x,y)
plt.title('Scatter sample')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```

```python
random.seed(0)
np_data_x = np.arange(1000)
np_data_y = np.random.randn(1000).cumsum()

plt.figure(figsize=(20,6))

plt.plot(np_data_x,np_data_y,label='Label')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```

```python
plt.figure(figsize=(20,6))
plt.subplot(2,1,1)
x = np.linspace(-10,10,100)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
y = np.linspace(-10,10,100)
plt.plot(y,np.sin(2*y))

plt.grid(True)
```

```python
def my_func(x):
    return x**2 + 2*x + 1

x = np.arange(-10,10)
plt.figure(figsize=(10,5))
plt.plot(x,my_func(x))
plt.grid(True)
```

```python
# 2章総合問題(2)(3)
plt.figure (figsize=(20,20))

x = np.random.uniform(0.0,1.0,10000)
y = np.random.uniform(0.0,1.0,10000)

cnt = 0
for i in np.arange(10000):
    if (np.hypot(x[i],y[i]) < 1):
        cnt += 1
print (cnt)
print (cnt*4/10000)
```
